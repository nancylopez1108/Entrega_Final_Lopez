**Proyecto : Asistente inteligente para recomendación de destinos turísticos personalizados **


1_**Resumen:**
Este proyecto utiliza un modelo de inteligencia artificial para recomendar destinos turísticos en Argentina según las preferencias del usuario, como presupuesto, duración del viaje y tipo de entorno deseado. El sistema primero genera recomendaciones de destinos mediante un prompt estructurado y luego muestra información e imágenes ilustrativas de cada lugar sugerido. De esta forma se simula un asistente turístico inteligente que ayuda a planificar viajes de manera personalizada.


**2_Introducción a la problemática:**
Planificar un viaje turístico suele ser una tarea compleja para muchas personas, especialmente cuando se deben tener en cuenta múltiples variables como el presupuesto disponible, el tiempo del que se dispone, el tipo de viaje deseado (descanso, aventura, cultural), la cantidad de personas o las preferencias personales.
En muchos casos, la falta de información clara en internet genera confusión y dificulta la toma de decisiones, provocando que las personas no sepan qué destino elegir o terminen optando por opciones que no se ajustan realmente a su deseo.
Este problema es relevante porque el turismo es una actividad vinculada al bienestar y el ocio de las personas. Contar con una herramienta que ayude a orientar la elección de un destino de manera personalizada puede mejorar significativamente la experiencia del usuario.


3_**Solución propuesta:**
La propuesta consiste en desarrollar un asistente basado en inteligencia artificial que recomiende destinos turísticos personalizados mediante la correcta generación de prompts, utilizando tanto modelos de texto-texto como de texto-imagen.

El sistema funcionará de dos maneras las cuales son complementarias:
1. Recomendación a partir de un cuestionario (texto → texto / texto → imagen)
El usuario responderá un cuestionario simple, que incluirá:
A)	Presupuesto aproximado
B)	Tiempo disponible para el viaje
C)Tipo de viaje (descanso, aventura, cultural)
D)Preferencia de entorno (playa, montaña, ciudad, naturaleza)
E)Cantidad de personas
F)Fechas  

Modelo texto-texto
Se diseñarán prompts que analicen las respuestas del cuestionario y generen recomendaciones de destinos acordes a los datos ingresados.
Ejemplo de prompt:
“Actuá como un asesor turístico. En base a los siguientes datos del viajero, recomendá uno o más destinos turísticos adecuados, explicando brevemente por qué se ajustan a sus preferencias.”.Para este modelo usaremos CHAT GPT o GEMINI

Modelo texto-imagen
A partir del destino recomendado, se generarán imágenes representativas del lugar, permitiendo al usuario visualizar la propuesta.
Ejemplo de prompt:
“Genera una imagen realista y atractiva de un destino turístico ideal para descanso, con paisaje natural y ambiente relajado.” En este modelo se utilizaran las mismas tecnologias.
